### Log transformation

```python
#Log tranformation
def image_negative (img):
    L=np.max(img)
    negative_image = L-img
    return negative_image

#r= is your image
def log_tranformation (r,c):
    r = r.astype(float)
    s = c *np.log(1+r)
    return s.astype(np.uint8)

def stack(*args):
    return np.hstack(args)

def rescale(img):
    s=img.astype(float)
    s-= np.min(s)
    s /= np.max(s)
    return (s*255).astype(np.uint8)

img = cv2.imread('../src/assets/current.png',0)
neg_image=image_negative(img)
log_foto =log_tranformation(img, c=1)
log_foto = rescale(log_foto)

yan_neg_img = stack(img, neg_image)
yan_logs_img = stack(img, log_foto)

plt.imshow(yan_logs_img, cmap="gray")
plt.show()
```
