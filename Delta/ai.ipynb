##### Camera

```python
cam = cv2.VideoCapture(0)

#lees parameter vanuit de camera
print(cam.get(3))
print(cam.get(4))

#Set parameter van de camera
cam.set(3,340)
cam.set(4,220)


while True:
    ret, frame = cam.read()
    frame=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)#Hiermee zorgen we swart wit beeld
    
    if not ret:
        print("camera not view")
        break
    
    cv2.imshow('Camera', frame)
    
    if cv2.waitKey(1) & 0xFF == ord("q"):
        print("view endet")
        break
cam.release()
cv2.destroyAllWindows()
```

##### Geometric draw

```python
img = np.zeros((512,512,3),np.uint8)

cv2.line(img,(0,0),(511,511),(255,0,0),2)
cv2.circle(img,(400,200),50,(0,255,0),-1)
cv2.rectangle(img,(500,20),(400,300),(255,0,0),2)
cv2.ellipse(img,(256,256),(100,50),0,0,180,(255,0,0),2)

font=cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img,'opencv',(20,400),font,4,(0,155,155),2,cv2.LINE_AA)

pts =np.array([[20,30],[100,120],[255,255],[10,400]],np.int32)
pts2 = pts.reshape(-1,1,2) # Begind vanuit index -1 neem bij elk rij twee kolomen

cv2.polylines(img,[pts],True,(255,255,255),1)# True sluit alle punten met elkaar en 3 is de dikte van je lijn

cv2.imshow("Demo ",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python


def foto_negatief(foto):
    L=np.max(foto)
    negatief_foto = L-foto
    return negatief_foto

img = cv2.imread('../src/assets/camera.png',0)
negatie_foto = foto_negatief(img)

yanyana=np.hstack((img,negatie_foto))
plt.imshow(yanyana,cmap="gray" )
plt.show()
```
